```python
# importing the liberary
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
lr=float(input("Enter the learning rate "))
j=int(input("Enter the initial point"))
def y_function(t):
    return (t-2)**2
def derivative(t):              #function to calculate the derivative
    x = sp.symbols('x')
    d = sp.diff((x-2)**2, x)
    m=d.subs(x,t)
    return m
x=np.arange(-100,100,0.1)
y=y_function(x)
```

```python
def gradient_descent():
     current_pos = (j,y_function(j))
     x_values = []  
     y_values = []
     iteration=100
     for i in range (iteration):                #main logic of gradient descent 
        new_x = current_pos[0] - lr * derivative(current_pos[0])
        new_y = y_function(new_x)
        x_values.append(new_x) 
        y_values.append(new_y)
        plt.plot(y_values)
        current_pos = (new_x, new_y)
        plt.plot(x,y,label='Gradient Descent Path')
        print(current_pos[0],current_pos[1])
        plt.scatter(current_pos[0],current_pos[1], color="red")    
        if abs(derivative(current_pos[0])) < 1e-6:
            print("optimum x and optimum y are...", new_x,new_y)
          
         
        
```

```python
gradient_descent()          #calling the function
```
